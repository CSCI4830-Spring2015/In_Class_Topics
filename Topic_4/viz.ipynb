###VnetCDF Class</a>

This is a basic framework for encapsulating the NetCDF module.

```python
''' This module contains the visualization encapsulation of a 
NetCDF file.'''
__all__=[] # prevent accidental "from viz.VnetCDF import *"

import numpy     as np
import netCDF4   as nc
import matplotlib.pyplot as plt

class VnetCDF(object):
    
    ''' This is a encapsulation class for a NetCDF file. To be expanded
    upon and used in the Visualization course. It will primarily wrap a
    single file.
    
    NOTE: You should call the close() method just before destroying the
          object. As this closes the file and will flush all data to disk.
    '''
    
    def __init__(self, filename=None):
        '''Create a new visualization NetCDF object.
        Arguments are:
            filename -- The NetCDF file name.
        '''
        self._filename = None
        self._ds       = None
        self._ndims    = None

        if filename:
            self._filename = filename
            try:
                self._ds = nc.Dataset(filename, mode="r+")
            except Exception as e:
                print "Error with %s: %s" %(filename, e)
                raise
                
    def __enter__(self):
        return self
    
    def __exit__(self, type, value, tb):
        self.close()
        
    
    def close(self):
        '''Close the NetCDF file.''' 
        if self._ds:
            self._ds.close()
            
    @property
    def filename(self):
        '''The NetCDF filename.'''
        return self._filename
  
    
#
# Functions to deal with global dimensions. 
#

    @property
    def ndims(self):
        '''The number of dimensions.'''
        self._ndims = len(self._ds.dimensions)
        return self._ndims
    
    def get_global_dim_names(self):
        '''Get the global dimension names, as a list.'''
        return self._ds.dimensions.keys()
    
    def get_global_dim_lens(self):
        '''Get the global dimension lenghts, as
        a dictionary with the dimension name as the key.'''
        return {k: len(self._ds.dimensions[k]) for k,v in self._ds.dimensions.items()}

#
# Functions to deal with variables.
#
    
    @property
    def nvars(self):
        '''The number of variables.'''
        return len(self._ds.variables)
    
    def get_var_names(self):
        '''Get the names of the variables.
        This will return a list of all the variables.'''
        return self._ds.variables.keys()
    
    def get_var_dim_lens(self, var):
        '''The dimensions of the variable var.'''
        return self._ds.variables[var].size
    
    def get_var_dim_names(self, var):
        '''The names for each dimension of the variable var.'''
        return self._ds.variables[var].dimensions 
    
    def get_num_of_points(self, var):
        '''The number of grid points where var is defined.'''
        data = self.get_data(var)
        try:
            missing = data.missing_value
            valid = np.ma.masked_values(data, missing)
            return np.ma.count(valid)
        except AttributeError:
            return np.prod(self._ds.variables[var].shape)
    
    def get_num_of_cells(self, var):
        '''The number of cells where var is defined.
        
        NOTE: This we should be using cf-python and friends.
              As the dataset confirms to the CF (Climate and Forecast)
              metadata (http://cfconventions.org/).
              That is the use of _bnds to describe the bounding box/cells.
        '''
        data = self.get_data(var)
        return np.prod([s -1 for s, d in zip(data.shape, data.dimensions) if d != 'time' ])
    
    def get_data(self, var):
        '''Get the data for variable var.'''
        return self._ds.variables[var]
    
    def plot_histogram(self, var):
        '''Plot a histogram of the variable var.'''
        bins = 20
        fig = plt.figure()
        plt.xlabel('Variable %s' %var)
        plt.ylabel('Counts')
        plt.title('Histogram from %s' %self._filename)
        
        data = self.get_data(var)
        try:
            missing = data.missing_value
            valid = np.ma.masked_values(data, missing)
            hdata = valid.compressed()
        except AttributeError:
            if data.ndim != 1:
                hdata = data.flatten()
            else:
                hdata = data
        plt.hist(hdata, bins)
        

        
```

###Unit Tests</a>

We will write a unit test for each function in the class.

Creating a test suite that runs all the tests.

```python
import unittest

class TestVnetcdf(unittest.TestCase):
    
    def setUp(self):
        self.vnetcdf = VnetCDF('tos_O1_2001-2002.nc')
        
    def test_ndims(self):
        expected = 4
        ndims = self.vnetcdf.ndims
        self.assertEqual(ndims, expected,
                         'incorrect number of dimensions')
        
    def test_get_global_dim_names(self):
        expected = [u'lon', u'lat', u'time', u'bnds']
        names = self.vnetcdf.get_global_dim_names()
        self.assertListEqual(names, expected,
                             'incorrect global dimension names')
    
    def test_get_global_dim_lens(self):
        expected = {u'lat': 170, u'bnds': 2, u'lon': 180, u'time': 24}
        lens = self.vnetcdf.get_global_dim_lens()
        self.assertDictEqual(lens, expected,
                             'incorrect global dimension names')
        
    def test_nvars(self):
        expected = 7
        nvars = self.vnetcdf.nvars
        self.assertEqual(nvars, expected,
                         'incorrect number of dimensions')
        
    def test_get_var_names(self):
        expected = [u'lon', u'lon_bnds', u'lat', u'lat_bnds', u'time', u'time_bnds', u'tos']
        names = self.vnetcdf.get_var_names()
        self.assertListEqual(names, expected,
                             'incorrect variable names')
    
    def test_get_var_dim_lens(self):
        expected = 180
        lens = self.vnetcdf.get_var_dim_lens('lon')
        self.assertEqual(lens, expected,
                             'incorrect variable dimension length')
    
    def test_get_var_dim_names(self):
        expected = (u'time', u'lat', u'lon')
        names = self.vnetcdf.get_var_dim_names('tos')
        self.assertTupleEqual(names, expected,
                             'incorrect variable dimension names')
        
    def test_get_num_of_points(self):
        expected = 506160
        npoints = self.vnetcdf.get_num_of_points('tos')
        self.assertEqual(npoints, expected,
                         'incorrect number of points')
    
    def test_get_num_of_cells(self):
        expected = 30251
        ncells = self.vnetcdf.get_num_of_cells('tos')
        self.assertEqual(ncells, expected,
                         'incorrect number of cells')
    
    def test_get_data(self):
        expected = -79.5
        point = self.vnetcdf.get_data('lat')[0]
        self.assertEqual(point, expected,
                         'incorrect data at lat[0]')    
    
def suite():
    tests = ['test_ndims',
             'test_get_global_dim_names',
             'test_get_global_dim_lens',
             'test_nvars',
             'test_get_var_names',
             'test_get_var_dim_lens',
             'test_get_var_dim_names',
             'test_get_num_of_points',
             'test_get_num_of_cells',
             'test_get_data']
    return unittest.TestSuite(map(TestVnetcdf, tests))
    
if __name__ == '__main__':
    runner = unittest.TextTestRunner()
    tests = suite()
    runner.run(tests)
```
